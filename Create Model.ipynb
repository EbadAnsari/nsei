```python
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("NSEI (Rules).csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
dateSplit = pd.to_datetime('01-01-2025', format="%d-%m-%Y")
df = df[df['Date'] < dateSplit]
df.index = range(len(df))
```

```python
x = df[['Open', 'High', 'Low', 'Close']]
y = df['classifier']
```

```python
lr = LogisticRegression()
lr.fit(x, y)
```

```python
df.head()
```

```python
tree = DecisionTreeClassifier()
tree.fit(x, y)
```

```python
svc = SVC()
svc.fit(x, y)
```

```python
knn = KNeighborsClassifier()
knn.fit(x, y)
```

```python
pd.DataFrame({
    "Model": ["Logistic Regression", "Decision Tree", "Support Vector Machine", "KNN"],
    "Accuracy": [accuracy_score(y, lr.predict(x)), accuracy_score(y, tree.predict(x)), accuracy_score(y, svc.predict(x)), accuracy_score(y, knn.predict(x))]
})
```
