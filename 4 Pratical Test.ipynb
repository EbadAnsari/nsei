```python
df = pd.read_csv('NSEI (Rules).csv')
```

```python
df.head()
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(df[["Open", "High", "Low", "Close"]], df["classifier"], test_size=0.7)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(xTrain, yTrain)

tree = DecisionTreeClassifier()
tree.fit(xTrain, yTrain)

knn = KNeighborsClassifier()
knn.fit(xTrain, yTrain)

svc = SVC()
svc.fit(xTrain, yTrain)
```

```python
accuracy_score(yTest, lr.predict(xTest))
```

```python
accuracy_score(yTest, tree.predict(xTest))
```

```python
accuracy_score(yTest, knn.predict(xTest))
```

```python
accuracy_score(yTest, svc.predict(xTest))
```

```python
comparative = pd.DataFrame(columns=["Accuracy", "F1", "Recall", "Precision"])
```

```python
comparative.loc["Logistic Regression"] = [accuracy_score(yTest, lr.predict(xTest)), f1_score(yTest, lr.predict(xTest)), recall_score(yTest, lr.predict(xTest)), precision_score(yTest, lr.predict(xTest))]
comparative.loc["Decision Tree"] = [accuracy_score(yTest, tree.predict(xTest)), f1_score(yTest, tree.predict(xTest)), recall_score(yTest, tree.predict(xTest)), precision_score(yTest, tree.predict(xTest))]
comparative.loc["KNN"] = [accuracy_score(yTest, knn.predict(xTest)), f1_score(yTest, knn.predict(xTest)), recall_score(yTest, knn.predict(xTest)), precision_score(yTest, knn.predict(xTest))]
comparative.loc["SVC"] = [accuracy_score(yTest, svc.predict(xTest)), f1_score(yTest, svc.predict(xTest)), recall_score(yTest, svc.predict(xTest)), precision_score(yTest, svc.predict(xTest))]
```
